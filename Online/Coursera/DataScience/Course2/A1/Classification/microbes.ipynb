```r
data <- read.csv(file="seaflow_21min.csv", header=TRUE, sep=",")
```

```r
data
```

```r
summary(object = data)
```

```r
# Split the data into test and training sets
set.seed(21)

half_size <- floor(nrow(data)/2)

rand <- data[sample(nrow(data)),]
training <- rand[1:half_size,]
test <- rand[(half_size+1):nrow(rand),]
```

```r
summary(training)
```

```r
library(ggplot2) 
qplot(pe, chl_small, data = data, color = pop)
```

```r
# Train decision tree
library(rpart)
fol <- formula(pop ~ fsc_small + fsc_perp + fsc_big + pe + chl_big + chl_small)
model_rpart <- rpart(fol, method="class", data=training)
print(model_rpart)
```

```r
# Evaluate decision tree
prediction <- predict(model_rpart, test, type='class')
summary(prediction == test$pop)
```

```r
30938/(30938+5234)
```

```r
# Random forest
library(randomForest)
model_rf <- randomForest(fol, method="class", data=training)
prediction <- predict(model_rf, test, type='class')
summary(prediction == test$pop)
```

```r
33293/(33293+2892)
```

```r
importance(model_rf)
```

```r
library('e1071')
model_svm <- svm(fol, method="class", data=training)
print(model_svm)
```

```r
prediction <- predict(model_svm, test, type='class')
summary(prediction == test$pop)
```

```r
33228/(33228+2944)
```
